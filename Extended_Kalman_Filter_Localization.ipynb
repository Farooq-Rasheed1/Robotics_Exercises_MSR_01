## 6.1 Prediction [3.0]

The `ekf_predict` function computes a prediction about the robot's pose after moving by using the odometry motion model.

It takes as input:

- the current belief about the pose of the robot represented as a Gaussian distribution $\mathcal{N}(\mu_t,\Sigma_t)$ 
- the odometry readings $u_t$

The output is a prediction about the robot's pose $\mathcal{N}(\overline{\mu}_{t+1},\overline{\Sigma}_{t+1})$.

You can use your implementation of the `inverse_motion_model` function from Exercise 3 to compute the $u_t = [\delta_{rot1}, \delta_{trans}, \delta_{rot2}]$  from the odometry information.

Implement the `ekf_predict` function and verify that it is correct for some test input. 

```python
# Test the prediction step
# Initialize test values
mu_test = np.array([1.0, 1.0, 0.0])  # Initial pose [x, y, theta]
S_test = np.eye(3) * 0.1  # Initial covariance
u_test = data['odom'][10]  # Get odometry reading at step 10
R_test = np.diag([0.1, 0.1, 0.05])  # Motion noise

# Run prediction
mu_pred, S_pred = ex.ekf_predict(mu_test, S_test, u_test, R_test)

print("Initial state:")
print("mu =", mu_test)
print("\nOdometry reading:")
print("u =", u_test)
print("\nPredicted state:")
print("mu_pred =", mu_pred)
print("\nPredicted covariance:")
print("S_pred =\n", S_pred)

```

## 6.2 Correction [3.0]

The `ekf_correct` implements the correction step of the EKF that corrects the prediction according to the sensor measurements.

It takes as input:

- the current prediction about the pose of the robot represented as a Gaussian distribution $\mathcal{N}(\overline{\mu}_{t+1},\overline{\Sigma}_{t+1})$
- the sensor measurements $z_t$

The output is new belief about the robot's pose $\mathcal{N}({\mu}_{t+1},{\Sigma}_{t+1})$.

Implement the `ekf_correct` function and verify that it is correct for some test input

```python
# Test the correction step with more realistic values
# Initial prediction state (simulating output from predict step)
mu_pred = np.array([1.0, 1.0, 0.0])  # Predicted pose
S_pred = np.array([[0.5, 0.0, 0.0],  # Larger initial uncertainty
                   [0.0, 0.5, 0.0],
                   [0.0, 0.0, 0.3]])

# Sensor measurements
z_test = np.array([
    [2.0, 0.5, 0],    # Range, bearing, landmark_id
])

# Smaller measurement noise (more trust in measurements)
Q_test = np.diag([0.1, 0.1])  # Measurement noise for [range, bearing]

# Get the landmark positions
M_test = M  # Using the map from your dataset

# Run correction
mu_corr, S_corr = ex.ekf_correct(mu_pred, S_pred, z_test, Q_test, M_test)

# Print results
print("Predicted state:")
print("mu_pred =", mu_pred)
print("\nPredicted covariance:")
print("S_pred =\n", S_pred)
print("\nCorrected state:")
print("mu_corr =", mu_corr)
print("\nCorrected covariance:")
print("S_corr =\n", S_corr)
print("\nUncertainty reduced by:", np.trace(S_pred) - np.trace(S_corr))

```
